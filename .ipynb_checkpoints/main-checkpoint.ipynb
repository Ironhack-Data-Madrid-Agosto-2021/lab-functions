## 2. Now write a function that returns the largest element on a list

```python
def greatest(arr):
    lar = arr[0]
    for ele in arr:
        if ele > lar:
            lar = ele
    return lar
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    fact = n
    while n > 1:
        fact = fact*(n-1)
        n = n-1
    return fact
        
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ðŸ¤”`

```python
def unique(arr):
    new_list = list()
    for i in arr:
        valida_valor_lista = False
        for y in new_list:
            if y == i:
                valida_valor_lista = True
                break
        if valida_valor_lista == False:
            new_list.append(i)
    return new_list
            
        
```

```python
# This will test your function 
test_unique(unique)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
    unico = unique(arr)
    my_dict = dict()
    for ind, ele in enumerate(unico):
        
        var = 0
        for ind2, ele2 in enumerate(arr):
            if ele == ele2:
                var += 1
        my_dict[var] = ele
    mode = my_dict[greatest(list(my_dict.keys()))]
    return mode
    
```

```python
# This will test your function 
test_mode(mode_counter)
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
from statistics import stdev

def leng(arr):
    value = sum_all([1 for x in arr])
    return value
```

```python
def st_dev(arr):
    media = sum_all(arr) / leng(arr)
    diff_sq = [(media - x) ** 2 for x in arr]
    rest = (sum_all(diff_sq) / (leng(arr)-1)) ** 0.5
    return rest
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    letras = 'abcdefghijklnmopqrstuvwxyz'
    set_letra = unique(string.lower())
    contador = 0
    for letra in set_letra:
        for letra2 in letras:
            if letra2 == letra:
                contador = contador + 1
                if contador == leng(letras):
                    return True
    return False
pangram("The quick brown fox jumps over the lazy dog.")
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
    new_str = string.split(',')
    new_str.sort()
    sorted_str = ','.join(new_str)
    return sorted_str
sort_alpha('hoy,hace,frio')
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.
`Valid special characters: # @ ! $ % & ( ) ^ * [ ] { }`

```python
def check_pass(string):
    valid_special_char = '# @ ! $ % & ( ) ^ * [ ] { }'
    a = 0
    b = 0
    c = 0
    d = 0
    e = 0
    if len(string) >=8:
        a += 1
    for x in string:
        if x.isupper():
            b += 1
        if x.islower():
            c += 1
        if x.isnumeric():
            e += 1
    for char in string:
        for char2 in valid_special_char:
            if char == char2:
                d += 1
    if a > 0 and b > 0 and c > 0 and d > 0 and e > 0:
        return True
    else:
        return False

print(check_pass('Aaaaa1a'))
            
        
    
```

```python
# This will test your function 
test_pass(check_pass)
```
